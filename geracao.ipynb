```python
from langchain.vectorstores import Chroma
from langchain_huggingface import HuggingFaceEmbeddings
from langchain.llms import Ollama
```

```python
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/LaBSE")
```

```python
vector_local = Chroma(
    collection_name="galo",   
    embedding_function=embeddings,
    persist_directory="./local_chroma_db",
)
```

```python
query = 'Quem são os ídolos do galo?'

resp = vector_local.similarity_search_with_score(query=query, k=4)

resp
```

```python
page_contents = [doc.page_content for doc, _ in resp]

page_contents
```

```python
from langchain.llms import Ollama

prompt = f"""
    Você é um assistente de perguntas e respostas sobre o Atlético Mineiro.
    Você deve responder as perguntas com base no contexto abaixo.
    Se não souber a resposta, diga que não sabe.
    Contexto: {page_contents}
    Pergunta: {query}
    Resposta:
"""
ollama = Ollama(model="llama3.1")
response = ollama(prompt)
print(response)
```
