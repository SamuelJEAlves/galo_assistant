```python
import chromadb
from langchain.vectorstores import Chroma
from langchain_huggingface import HuggingFaceEmbeddings
```

```python
# Exemplo da possibilidade de usar a biblioteca oficial do ChromaDB para algumas operações

persist_directory = "local_chroma_db"

client = chromadb.PersistentClient(path=persist_directory)

print(client.heartbeat())

client.list_collections()
```

```python
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/LaBSE")
```

```python
vector_local = Chroma(
    collection_name="galo",   
    embedding_function=embeddings,
    persist_directory="./local_chroma_db",
)
```

```python
vector_local._collection
```

```python
query = 'Quem são os ídolos do Atlético Mineiro?'

resp = vector_local.similarity_search_with_score(query=query, k=4)

resp
```
